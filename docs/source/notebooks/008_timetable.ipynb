# Timeplan for kurs i Mars 2026 OsloMet

## Forutsetninger
1) Alle kursdeltakere har en fungerende og testet konto på Educloud on Demand. Følg instruksjoner her [here](notebooks/01_easy_login.ipynb).

2) Carpentry leksjonen on [Plotting and Programming in Python](
https://swcarpentry.github.io/python-novice-gapminder) bør tas som et selvstudium. Som et minimum kapitlene en og seks. Alternativt boken [Databehandling og -analyse med Python og OpenRefine](https://python4abm.github.io/python4abm/intro.html), bare kapitlene en til tre.

3) Finn skjermtastaturet på maskinen du skal bruke. Du kommer til å bli glad for at du vet hvor det er, når vi er i gang.

## Alle må ta med
* Bærbar maskin
* Smarttelefon med [Autentisering](https://www.microsoft.com/nb-no/security/mobile-authenticator-app). Testet og klar til bruk i Educloud on demand.
