Title: **Important** :Required preparations for "Run LLMs on Educloud UiO", Digital Scholarship days

Hi Everyone,

Welcome to Digital Scholarship Days 2026 and Run Large Language Models on Educloud UiO.

There are important preparations for our course wich is held as part of Digital Scholarship days, Run LLMs on Educloud UiO.

In order to follow the course, it is essential that everyone has logged into Educloud. This is where we are going to work. The url is: `https://ood.educloud.no<https://ood.educloud.no>`_.

In order to enter, you need the Url for signing up and `applying for group membership <https://selfservice.educloud.no/membership-application>`_.

We need all of you to apply for group membership in ec443. This is why we have made `this thorough guide<https://llm-course-eng.readthedocs.io/en/latest/notebooks/01_preparations.html>`_.

We will make room for meeting with those of you who have questions regarding these practical issues, right after the x-mas holiday. But before that, please try it on your own!
Please do not hesitate sending me an email, if you have comments on this procedure, as i look forward to reading some email also during the holiday.

Have a very merry x-mas to everyone!

Best wishes from the LLM-team
