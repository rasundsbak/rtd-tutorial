### Title: **Viktig** :Obligatoriske forberedelser til "Kjør store språkmodeller på UiO Educloud".

Hei Alle,

Velkommen til kurset Kjør store språkmodeller på UiO Educloud.

Her kommer informasjon om noen viktige forberedelser til kurset. For at man skal henge med, bør man ha øvd på innlogging i Educloud. Educloud har en nettleserversjon. Her er url for innlogging: [https://ood.educloud.no ](https://ood.educloud.no).

For at dette skal virke, må du først [søke om gruppemedlemsskap](https://selfservice.educloud.no/membership-application).

Gruppen heter: **ec443**. For å gjøre prosessen enkel, har vi laget [denne grundige gjennomgangen](https://rtd-2-tutorial.readthedocs.io/no/latest/notebooks/26_preparations.html).

Ved behov, vil jeg sette av tid til å treffe de av dere som har spørsmål rundt dette, i et Zoom møte, slik at alle er inne og jobber på rett sted når kurset starter. Jeg er selvsagt også klar på epost, og vil svare ut alle spørsmål dere måtte ha.

Vi anbefaler at dere som deltar gjør forberedelser. Se mer om dette ved [timeplanen for kurset](https://rtd-2-tutorial.readthedocs.io/no/latest/notebooks/008_timetable.html):

Carpentry leksjonen om [Plotting and Programming in Python](https://swcarpentry.github.io/python-novice-gapminder/) bør tas som et selvstudium. Som et minimum kapitlene en og seks. Alternativt boken [Databehandling og -analyse med Python og OpenRefine](https://python4abm.github.io/python4abm/intro.html), bare kapitlene en til tre.
